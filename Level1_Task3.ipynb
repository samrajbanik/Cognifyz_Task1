```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/SAMRAJ/Downloads/Cognifyz Dataset.csv")
```

```python
data.columns
```

```python
# Plot a bar chart for the distribution of price ranges
price_range_counts = data['Price range'].value_counts()
price_range_counts.plot(kind='bar', color='skyblue')

plt.title('Distribution of Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.show()

# Calculate the percentage of restaurants in each price range category
total_restaurants = len(data)
percentage_price_ranges = (price_range_counts / total_restaurants) * 100

print("\nPercentage of restaurants in each price range category:")
print(percentage_price_ranges)
```
